```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib import rc
rc('text', usetex=True)
%matplotlib inline
```

## For fixed inputs

```python
t = [1,2,3,4,5,6,8,9,10,12,15,18,20,24,30,36,40,45,60]

y = [51.288480,25.817611,19.441015,13.326728,11.786567,9.501848,7.509774,7.775930,6.327702,5.551481,5.096534,
4.052896,3.709848,3.108163,2.463560,2.092774,1.947679,1.775563,1.273707]
```

```python
fig, ax = plt.subplots(figsize=(10,7))

ax.plot(t, y, alpha=0.8, linewidth=1)

ax.set_xlabel("Size", fontsize=18)
ax.set_ylabel("Runtime", fontsize=18)

ax.set_xlim(t[0], t[18])
ax.set_ylim(y[18],y[0])

ax.xaxis.set_major_locator(MultipleLocator(5))
#ax.xaxis.set_major_formatter('{x:.0f}')
#ax.xaxis.set_minor_locator(MultipleLocator(3))
ax.yaxis.set_major_locator(MultipleLocator(5))
#ax.yaxis.set_major_formatter('{x:.0f}')
#ax.yaxis.set_minor_locator(MultipleLocator(3))

for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(13)
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(13)

ax.grid(which='major')
ax.grid(which='minor', alpha=0.5)

fig.tight_layout()

fig.savefig('fixedinput_figure.eps')

plt.show()
```
